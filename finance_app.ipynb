```python
import pandas as pd
import os
```

```python

# Set Directory Paths/Keys
#os.chdir('~/Box Sync/Summer 2016/finance_app/')
dl_dir = '~/Downloads/transactions.csv'
out_dir = '~/Dropbox/Finance_App.xlsx'

# Todo: Autodownload

# Read Data in, index as datetime indices
df = pd.read_csv(dl_dir, parse_dates = 'Date', index_col = 'Date')

#df = df.truncate(after = '07/01/2016')

# Defining Groups

food = ['Restaurants','Fast Food','Groceries','Food & Dining']
coffee = ['Coffee Shops']
transport = ['Public Transportation','Gas & Fuel', 'Rental Car & Taxi','Auto & Transport','Parking']
travel = ['Travel','Coach Bus','Air Travel','Hotel']
drinks = ['Alcohol & Bars']
sport = ['Sports','Sporting Goods','Hobbies']
shopping = ['Shopping','Electronics & Software','Music','Books','Clothing','Books & Supplies','Gift','Office Supplies','Home Improvement','Mobile Phone'] # should it be a set?
entertainment = ['Event Tickets','Entertainment','Amusement','Arts']
fees = ['ATM Fee','Finance Charge','Fees & Charges','Bank Fee','State Tax','Federal Tax','Service Fee']
misc = ['Printing','Hair','Uncategorized','Business Services','Transfer for Cash Spending','Doctor','Eyecare','Misc Expenses', 'Cash & ATM']
home = ['Home','Utilities','Mortgage & Rent','Rent',]
edu = ['Education','Tuition']
hidden = ['Hide from Budgets & Trends']
allcats = set(food+coffee+transport+travel+drinks+sport + shopping+ entertainment+fees+misc+home+edu+hidden)
```

```python
# Find the difference between custom and actual categories
set(df.Category.unique()).difference(allcats)
```

```python
# Hardcode define dates/weeks. Week begins on Monday
wk1 = pd.date_range('7/6/2016', periods = 7, frequency = 'D')

# Last week
wk9 = pd.date_range('8/1/2016', periods=7, frequency = 'd')
wk10 = pd.date_range('8/8/2016', periods=7, frequency = 'd')
```

```python
# Use df.ix[date_range] 
df.ix[wk1]

# Filter: Pending, Category = hidden, Reimburseable?, tuition, rent
df[~df.Category.isin(hidden)].ix[wk1]
```

```python
# For each week we want to see: Category Spend vs. Budget vs. Historical
# TODO: Define budget

filtered_df = df.ix[(~df.Category.isin(hidden))&(~df.Category.isin(home))&(~df.Category.isin(edu))].ix[wk9]
```

```python
# Data
exp_food = filtered_df[filtered_df.Category.isin(food)]['Amount'].sum()
exp_coffee = filtered_df[filtered_df.Category.isin(coffee)]['Amount'].sum()
exp_transport= filtered_df[filtered_df.Category.isin(transport)]['Amount'].sum()
exp_travel= filtered_df[filtered_df.Category.isin(travel)]['Amount'].sum()
exp_drinks= filtered_df[filtered_df.Category.isin(drinks)]['Amount'].sum()
exp_sport= filtered_df[filtered_df.Category.isin(sport)]['Amount'].sum()
exp_shopping= filtered_df[filtered_df.Category.isin(shopping)]['Amount'].sum()
exp_entertainment= filtered_df[filtered_df.Category.isin(entertainment)]['Amount'].sum()
exp_fees= filtered_df[filtered_df.Category.isin(fees)]['Amount'].sum()
exp_misc= filtered_df[filtered_df.Category.isin(misc)]['Amount'].sum()
```

```python
spend_df = pd.Series(data=[exp_food,exp_coffee,exp_transport, exp_travel, exp_drinks, exp_sport, exp_shopping, exp_entertainment, exp_fees, exp_misc], index=['food','coffee', 'transport', 'travel', 'drinks', 'sport', 'shopping', 'entertainment','fees','misc'])
```
