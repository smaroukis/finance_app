```python
# Read in main
df = pd.read_csv('transactions.csv', parse_dates=['Date'], index_col = 'Date').truncate(after='10/1/2016')
```

```python
# Need to find flights, from a few weeks back (Category:=Air Travel in October)
flights = df[df.Category=='Air Travel'];
```

```python
# Begin and End date
sf = df.truncate(after = '10/23/16')
sf = sf.truncate(before = '10/28/16')
sf = pd.concat([sf,flights]) # Combine
null_labels = ['Hide']; null_cats = ['Utilities','Hide from Budgets & Trends']
sf = sf[~(sf.Category.isin(null_cats)|(sf.Labels.isin(null_labels)))]
```

```python
# Instead of using df[~df.Cat.isin(List)], to use regexp need pattern and str.contains
null_descr = ['iTunes','Airbnb*', 'Mighty Good','Jimmy John\'s','Mujo*', 'Continental', 'Panch*']
pat_null_descr = '|'.join(null_descr)
sf = sf[~sf.Description.str.contains(pat_null_descr)] # winner
```
